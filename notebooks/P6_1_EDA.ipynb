## Imports

```python
import pandas as pd
import numpy as np
import glob
import os
import chardet
import sys
sys.path.append('../lib')
from jbtools import describoost
```

## read_csv + describoost

```python
for filepath in glob.glob('../input/*.csv'):
    if os.path.basename(filepath) in ['data_clean.csv', 'sample_submission.csv', 'HomeCredit_columns_description.csv']: continue
    with open(filepath, 'rb') as f:
        encoding = chardet.detect(f.read(10000))['encoding']
    df2check = pd.read_csv(filepath, encoding=encoding)
    describoost(df2check, filepath)
```
